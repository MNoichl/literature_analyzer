### WOS-Data to t-SNE and Clustering

Put WebOfScience plaintext full records into the "data" folder, and the code below produces a t-SNE & Cluster-Analysis of cocitations.

```python

import metaknowledge as mk
import pandas as pd
import numpy as np


%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_context('poster')
sns.set_style('white')
sns.set_color_codes()
plot_kwds = {'alpha' : 0.35, 's' : 30, 'linewidths':0}
```

```python
RC = mk.RecordCollection("data")
```

```python
print(RC.glimpse())
```

```python
d = []
for R in RC:
    try:
        d.append((list(set(R.getCitations().get("author"))),R['id']))
    except KeyError:
        pass

df = pd.DataFrame(d)

df[0] = df[0].apply(lambda y: np.nan if len(y)<=2 else y) #here we kick out every paper that cites <= 2 authors
df = df.dropna()



authorslist = [';'.join(filter(None,x)) for x in list(df[0])]

#print(authorslist)

#print(pd.DataFrame(authorslist))

from sklearn.feature_extraction.text import CountVectorizer
import re

vec = CountVectorizer(token_pattern=r'(?<=[^|;])[\s\w]+(?=[$|;])')
X = vec.fit_transform(authorslist)
#pd.DataFrame(X.toarray(), columns=vec.get_feature_names())
```

```python
from sklearn.decomposition import PCA

PCA =  PCA(n_components=90) 
XPCA = PCA.fit_transform(X.toarray())
dPCA = pd.DataFrame(XPCA)
#print(dPCA[[0,1]])

```

```python
plt.scatter(dPCA[0], dPCA[1], color='b', **plot_kwds)
```

```python
from sklearn.metrics.pairwise import pairwise_distances

XCOS = pairwise_distances(XPCA, metric='cosine')
```

```python
from sklearn.manifold import TSNE


TSNE = TSNE(n_components=2, metric='precomputed', perplexity=16)#random_state=10
XTSNE = TSNE.fit_transform(XCOS)
dTSNE = pd.DataFrame(XTSNE)
plt.scatter(dTSNE[0], dTSNE[1], color='grey', **plot_kwds)
```

And now for the clustering.

```python
#import hdbscan
#clusterer = hdbscan.HDBSCAN(min_cluster_size=5, gen_min_span_tree=True)
#clusterer.fit()


from sklearn.cluster import DBSCAN

clusterer = DBSCAN(eps=0.9, min_samples=20).fit(XTSNE)
XCLUST = clusterer.labels_
#print(XCLUST)

color_palette = sns.color_palette('deep', 16)
cluster_colors = [color_palette[x] if x >= 0
                  else (0.5, 0.5, 0.5)
                  for x in clusterer.labels_]

plt.scatter(dTSNE[0], dTSNE[1], s=30, linewidth=0, c=cluster_colors, alpha=0.25)
```

```python
from sklearn.cluster import MiniBatchKMeans
clusternum=4
clusterer = MiniBatchKMeans(n_clusters=clusternum).fit(XTSNE)
XCLUST = clusterer.labels_

color_palette = sns.color_palette('deep', 16)
cluster_colors = [color_palette[x] if x >= 0
                  else (0.5, 0.5, 0.5)
                  for x in clusterer.labels_]

plt.scatter(dTSNE[0], dTSNE[1], s=30, linewidth=0, c=cluster_colors, alpha=0.45)

```

```python
dfclust = pd.DataFrame(XCLUST)
dfclust = pd.concat([df[1],dfclust],axis=1)
dfclust.columns = ['id','cluster']

drc = pd.DataFrame.from_dict(RC.forNLP(extraColumns=['journal']))

```
